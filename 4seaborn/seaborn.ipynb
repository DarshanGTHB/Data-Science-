```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
(data)
```

```python
# scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
sns.scatterplot(data=data, x='total_bill', y='tip')
# x=data['total_bill'] 
plt.subplot(1, 3, 2)
# sns.barplot(data=data, x='total_bill', y='tip')
# y=data['tip']
plt.subplot(1, 3, 3)
sns.lineplot(data=data, x='total_bill', y='tip')
# plt.scatter(x, y)
```

```python
sns.lineplot(data=data, x='size', y='tip')
```

```python
# catagorical plots

sns.barplot(data=data, x='day', y='total_bill' )
plt.show()
```

```python
# Box plot
print(data)
sns.boxplot(data=data, x='day', y='total_bill')
```

```python
sns.set_theme()
sns.violinplot(data=data, x='day', y='total_bill')
```

```python
sns.histplot(data['total_bill'], kde=True)
```

```python
sns.kdeplot( data['total_bill'], fill=True, color='purple')
```

```python
# PairPlot

sns.pairplot(data) # all pairs possible between "Numerical Columns"
```

```python
# Heatmap

corr = data[['total_bill', 'tip', 'size']].corr()
corr


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample correlation matrix

```

```python
sns.heatmap(corr, cmap='coolwarm', annot=True)
```

```python

# Get all colormaps
colormaps = plt.colormaps()
print(len(colormaps))
colormaps = colormaps[:8]
# Grid dimensions

plt.figure(figsize=(13,10))
plt.subplot( 2, 2, 2 )
sns.heatmap(corr, cmap='viridis')
# Plot each cmap in
for id, i in enumerate(colormaps):
    plt.title(i)
    plt.subplot( 4, 2, id+1 )
    sns.heatmap(corr, cmap=i)
```
